```typescript
async function clipToImage(text: string) {
  const resp = await fetch(
    `http://localhost:3001/clippify?text=${encodeURIComponent(text)}`
  );

  const image = await resp.arrayBuffer();
  // Must convert to base64 for Jupyter to display
  let imageForJupyterB = btoa(
    new Uint8Array(image).reduce(
      (data, byte) => data + String.fromCharCode(byte),
      ""
    )
  );

  Deno.jupyter.display({ "image/png": imageForJupyterB }, { raw: true });
}
```

```typescript
for (var word of ["hello", "world", "dino"]) {
  Deno.jupyter.display(word);
  await clipToImage(word);
}
```
